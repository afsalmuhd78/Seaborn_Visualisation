```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.columns
```

# LINE PLOT

```python
sns.lineplot(x = "SepalLengthCm", y= "PetalLengthCm", data = df)
plt.show()
```

```python
sns.lineplot(x ="SepalLengthCm", y= "PetalLengthCm", hue= "Species", data = df)
plt.show()
```

```python
sns.lineplot(x ="SepalLengthCm", y= "PetalLengthCm", hue= "Species",style = "Species", data = df)
plt.show()
```

```python
sns.lineplot(x= "SepalLengthCm", y="PetalLengthCm",hue= "Species", style="Species",markers=True, data= df)
plt.show()
```

# BAR PLOT

must have one axis numeric and another axis categorical,or two numeric,based on numeric values orientation changes


```python
sns.barplot(x="Species", y = "SepalLengthCm",data= df)
plt.show()
```

```python
sns.barplot(x="PetalWidthCm",y= "PetalLengthCm", hue="Species",data=df)
plt.show()
```

```python
data = pd.read_csv("Data.csv")
data.head()
```

```python
data.columns
```

```python
sns.barplot(x="Salary", y="Purchased", data=data)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm",hue="Species",data=df)
plt.show()
```

# HISTOGRAM / DISTPLOT
histogram is used to find the distribution of a continious numerical value


distplot used to plot histogram and frequency plot togethor

```python
sns.distplot(data["Salary"])
plt.show()
```

```python
sns.distplot(data["Salary"],hist=False)
plt.show()
```

```python
data["Salary"].mean()
```

```python
sns.distplot(data["Salary"],color="green",bins=5)
plt.show()
```

```python
sns.distplot(data["Salary"],color="red",kde=False,bins=5)
plt.show()
```

```python
sns.distplot(data["Salary"],color="black",vertical=True)
plt.show()
```

# JOINT PLOT

```python
sns.jointplot(x="SepalLengthCm",y= "SepalWidthCm",hue="Species",data=df)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm",kind="reg",data=df)
plt.show()
```

# B O X P L O T

```python
sns.boxplot(x="Purchased",y="Salary",data=data)
plt.show()
```

```python
sns.boxplot(x="Purchased", y="Age",data=data,linewidth=3)
plt.show()
```

```python
sns.boxplot(x="Country", y="Age",data=data,palette="Set1",order=["Germany","France","Spain"])
plt.show()
```

```python
sns.boxplot(x="Country",y="Salary",data=data,palette="Set1",order=["France","Germany","Spain"],hue="Purchased")# how categorical value change with numerical value
plt.show()
```

```python
s1= sns.countplot(data.Country, hue=data.Purchased)
s1.set_xticklabels(s1.get_xticklabels(),rotation=90)
plt.show()
```

```python
s2= sns.countplot(data.Age, hue=data.Purchased)
s2.set_xticklabels(s2.get_xticklabels(),rotation=90)
plt.show()
```

```python
s3= sns.countplot(data.Salary, hue= data.Purchased)
s3.set_xticklabels(s3.get_xticklabels(),rotation=90)
plt.show()
```
